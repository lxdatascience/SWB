```python
# drop the grand total line
df_lan = df_lan.dropna(subset=['League ID'])
```

```python
col_qstn = df_lan.columns[df_lan.columns.str.startswith('col')]
col_qstn
```

```python
# collect all Q & A's  
melted_df = df_lan.melt(id_vars=['League ID'], value_vars=col_qstn, var_name='Col_ID', value_name='Answer')
display(melted_df.head())
```

```python
# extract Question # and create Col_ID from metadata df to match with content df
info_df_lan['QstnNo'] = info_df_lan['Column'].str.extract(r'\b(\d{1,2})\.')
info_df_lan['Col_ID'] = 'col_' + info_df_lan.index.astype(str)
info_df_lan[['Col_ID', 'QstnNo', 'Column']]
```

```python
# extract Q&A related content only, and fill the only record (Row Index 11) with missing Q# with Qstn 4
info_qstn_lan = info_df_lan.iloc[6:,:].copy()
info_qstn_lan['QstnNo'] = info_qstn_lan['QstnNo'].fillna(method='ffill')

info_qstn_lan
```

```python
info_qstn_lan.info()
```

```python
# Create sub question seq # - QstnNo2 to store the sequence number of the answer choices for each question
info_qstn_lan['QstnNo2'] = info_qstn_lan['QstnNo'].str.strip() + ' ' + (info_qstn_lan.groupby('QstnNo').cumcount() + 1).astype(str)
info_qstn_lan.head(30)
```

```python
# Merge qstn_lan with info_qstn_lan to append the Question text to the answers
qstn_lan = pd.merge(melted_df, info_qstn_lan[['Col_ID', 'Column', 'QstnNo', 'QstnNo2']], on='Col_ID', how='left')
qstn_lan = qstn_lan.rename(columns={'Column': 'Qstn'})
qstn_lan['QstnNo'] = qstn_lan['QstnNo'].astype(int)
qstn_lan = qstn_lan[['League ID', 'Col_ID', 'QstnNo', 'QstnNo2', 'Qstn', 'Answer']]
qstn_lan.head()
```

```python
#detect missing/refused/other notes
def extract_answer_choice(qstnNo2):
    # QstnNo2=='7 1'
    qstn = qstn_lan.loc[qstn_lan.QstnNo2==qstnNo2, ['League ID', 'QstnNo2', 'Answer']].copy().reset_index(drop=True)
    
    Answer_Choices = qstn.Answer.str.split(';', expand=True)
    qstn = qstn.join(Answer_Choices)
    
    Answer_Choices_Cols = Answer_Choices.columns
    
    melted_qstn = qstn.melt(id_vars=['League ID'], value_vars=Answer_Choices_Cols, var_name='Choice_ID', value_name='Answer_Choice')
    
    display(melted_qstn['Answer_Choice'].value_counts().sort_values(ascending=False))
    
    display(qstn.loc[qstn[0].notna(), 'League ID'].nunique())
```
